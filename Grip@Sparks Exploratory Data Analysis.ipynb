### Let us first import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Let's import the data

```python
dataset=pd.read_csv('SampleSuperstore.csv')
dataset
```

### Let us analyse the data

```python
# checking the datatypes present in the dataset
dataset.dtypes
```

```python
#Check for any missing data in our dataset
dataset.isna().sum()
```

No Missing dataset is present in the dataset

```python
#Plotting a line plot between "City" And "Profit"
dataset.plot(x="City",y="Profit")
```

```python
fig,ax=plt.subplots(nrows=2,
                  ncols=2,
                  figsize=(10,10))
ax[0,0].plot(dataset["Category"],dataset["Profit"])
ax[0,1].scatter(dataset["Category"],dataset["Profit"])
ax[1,0].bar(dataset["Category"],dataset["Profit"])
ax[1,1].scatter(dataset["Quantity"],dataset["Profit"]);
```

```python
dataset.plot.hist(figsize=(10,30),bins=70,subplots=True);
```

```python
plt.style.use("dark_background")
#Customizing a plot with object oriented method of our SampleSuperstore dataset
fig,ax=plt.subplots(figsize=(15,15))
#creating a scattered plot
scatter=ax.scatter(x=dataset["Sales"],
                   y=dataset["Profit"],
                   c=dataset["Discount"],
                  cmap="winter")         #This changes the colour scheme
#naming the figure,x-axis,y-axis
ax.set(title="SampleSuperstore",
       xlabel="Sales",
       ylabel="Profit")
#to provide title and scale to the data by adding a legend 
ax.legend(*scatter.legend_elements(),title="Slope")
#adding a partition line
ax.axhline(dataset["Profit"].mean(),linestyle="--");
```

By Analysing various columns present in our dataset it gives a clear picture of how things are going and what needs to be changed in future to increase the Profit margins
